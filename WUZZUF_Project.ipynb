```python
#1st imports modules
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
```

```python
job_title = []
company_name = []
location = []
job_skill = []
links = []
salary =[]
ll = []
```

```python
#2nd step use requests to fetch the url
result = requests.get("https://wuzzuf.net/search/jobs/?q=python&a=hpb")
```

```python
#3rd step save page content
scr = result.content
print(scr)
```

```python
soup = BeautifulSoup(scr,"html")
print(soup)
```

```python
#5th step the elements containing info we need
#job titles, job skills, company names, location names
job_titles = soup.find_all('h2',{"class":"css-m604qf"} )
company_names = soup.find_all('a',{"class":"css-17s97q8"} )
locations = soup.find_all('span',{"class":"css-5wys0k"})
job_skills = soup.find_all('div',{"class":"css-y4udm8"})
#print()
```

```python
#6th step loop over retuned lists to extract needed info into other lists
for i in range(len(job_titles)):
    job_title.append(job_titles[i].text)
    links.append(job_titles[i].find("a").attrs['href'])
    company_name.append(company_names[i].text)
    location.append(locations[i].text)
    job_skill.append(job_skills[i].text)
print(links)


```

```python
for link in links:
    url = "https://wuzzuf.net"
    ul = url + link
    print(ll)
    result = requests.get(ll)
    scr = result.content
    scr
    soup = BeautifulSoup(scr,"html")
    salaries = soup.find("span" , {"class" :"css-4xky9y"})
    print(salaries)
    
    salary.append(salaries)
   
```

```python
#7th step create a new csv file and fill it with values
file_list = [job_title,company_name,location,job_skill,links,salary]
exported = zip_longest(*file_list) #unbacking
with open("jobs_test.csv","w") as myfile:
    wr = csv.writer(myfile)
    wr.writerow(["job title","company name","location","skills","links","salary"])
    wr.writerows(exported)
```
